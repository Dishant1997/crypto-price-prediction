```python
import nltk
import pandas as pd
nltk.download('vader_lexicon')
from pymongo import MongoClient
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
## mongoDB implementation
# build a new client instance for MongoClient
mongo_client = MongoClient('localhost', 27017)

# create new database and client collection
db = mongo_client.crypto
col = db.bitcoin_tweets

df = pd.DataFrame(list(col.find()))
df.head()
```

```python


df = pd.read_csv("bitcoin_tweets.csv")
df.head()
```

```python
df = df.drop(columns = ["Emojis"])
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df['scores'] = df['Text'].apply(lambda review: sid.polarity_scores(review))
df.head()
```

```python
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])
df = df.drop(columns = ["scores"])
df.head()
```

```python
df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
df.head()
```

```python
pos_count = 0
pos_score = 0.0
neg_count = 0
neg_score = 0.0
for index, row in df.iterrows():
    if row['comp_score'] == "pos":
        pos_count = pos_count + 1
        pos_score = pos_score + (row['compound']*float(row['Comments'][:-1]))
    elif row['comp_score'] == "neg":
        neg_count = neg_count + 1
        neg_score = neg_score + ((-1*row['compound'])*float(row['Comments'][:-1]))
```
